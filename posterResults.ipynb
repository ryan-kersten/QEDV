```python
import unittest
import QEDV_lib.error_codes
import QEDV_lib.verifier
from time import perf_counter
from matplotlib import pyplot as plt
import pandas as pd
```

```python
surface = QEDV_lib.error_codes.SurfaceCode(7)
print("Testing surface code of legnth 3")
print("X stabs are")
print(surface.x_stab)
print("Y stabs are")
print(surface.z_stab)
# val = QEDV_lib.verifier.minDistance(surface, 2)
# self.assertTrue(val)
# print(QEDV_lib.verifier.distSearch(surface))
# print(QEDV_lib.verifier.minDistance(surface,7))
args = [3,7]
timings = []
for i, arg in enumerate(args, 1):
    start = perf_counter()
    QEDV_lib.verifier.distSearch(QEDV_lib.error_codes.SurfaceCode(arg))
    end = perf_counter()
    duration = end - start
    timings.append({"Run": f"Surface Code d={2*i+1}", "Time (s)": duration})
df = pd.DataFrame(timings)
plt.figure(figsize=(8, 5))
plt.bar(df["Run"], df["Time (s)"], color="skyblue")
plt.xlabel("Run")
plt.yscale("log")  # Use log scale on y-axis
plt.ylabel("Execution Time (seconds)")
plt.title("Execution Time for Different Code Sizes")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig("execution_time_plot.png")  # Save to file



```
